```python
from transformers import WhisperForConditionalGeneration
import torch
import os
import shutil

#from google.colab import drive
#drive.mount('/content/drive')

# dir_path = '/content/drive/MyDrive/ML/DataSets/HindiSpeechRecognition/finalbatch2511'
# os.makedirs(dir_path, exist_ok=True)

model_dump = WhisperForConditionalGeneration.from_pretrained("CKSINGH/whisper-small-hi-finalBatch-test")



# Assume 'model' is your trained PyTorch model
# model = ...

# Specify the filename for saving the model
filename = 'pytorch_model_final_test.bin'

# Save the model state
torch.save(model_dump.state_dict(), filename)

print(f"Model saved as {filename}")


```

```python

# Define the source and destination file paths
source_path = '/content/pytorch_model_final_test.bin'
destination_path = '/content/drive/MyDrive/ML/DataSets/HindiSpeechRecognition/finalbatchnew2511/pytorch_model_final_test.bin'

# Copy the file
shutil.copy(source_path, destination_path)

print(f"File copied from {source_path} to {destination_path}")
```
